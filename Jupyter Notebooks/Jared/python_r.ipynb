```python
import pandas as pd
```

```python
df = pd.read_excel('data/sentiment.xlsx')
df = df.drop(['Unnamed: 0'], axis=1)
df
```

```python
emotions_df = pd.read_json('data/emotions.json')
emotions_df = emotions_df.apply(lambda x: x.str[0])
emotions_df
```

```python
df = df.merge(emotions_df, on='id')
df = df.set_index('id')
df
```

```python
df.to_excel('data/sentiment_emotions.xlsx')
```
